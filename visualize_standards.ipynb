```python
import pandas as pd
df = pd.read_csv("data/ASDIV_final.csv")
df
```

```python
len(df['substandard_VA'].unique())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# Set global font to Times New Roman and increase font size
mpl.rcParams['font.family'] = 'Times New Roman'
mpl.rcParams['font.size'] = 18  # You can increase this as needed
# Get value counts
value_counts = df['substandard_VA'].value_counts()

# Convert to DataFrame
value_counts_df = value_counts.reset_index()
value_counts_df.columns = ['substandard_VA', 'count']

# Start index at 1
value_counts_df['index'] = range(1, len(value_counts_df) + 1)

# Plot
plt.figure(figsize=(10, 6))
plt.bar(value_counts_df['index'], value_counts_df['count'])

# Use index numbers starting at 1 for x-ticks
plt.xticks(value_counts_df['index'], value_counts_df['index'])

plt.xlabel('Standard Number')
plt.ylabel('Count')
plt.title('Count of Each Unique Standard in ASDIV')
plt.tight_layout()
plt.savefig('asdiv_standard_counts.pdf')
plt.show()
```
